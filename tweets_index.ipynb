## Data Cleaning & Preprocessing

---

This section focuses on preparing the dataset for NLP and machine learning. We will handle missing values and duplicates, clean the tweet text by removing noise (URLs, mentions, hashtags, special characters), and create simple features like tweet length and word count. These steps ensure the data is structured, consistent, and ready for sentiment analysis.

```python
# -----------------------------
# Data Cleaning & Preprocessing
# -----------------------------

# Copy dataset
tweet_clean = tweet.copy()

# -----------------------------
# 1. Clean column names
# -----------------------------
tweet_clean.columns = tweet_clean.columns.str.strip().str.lower().str.replace(' ', '_')
tweet_clean.rename(columns={
    'is_there_an_emotion_directed_at_a_brand_or_product': 'sentiment',
    'emotion_in_tweet_is_directed_at': 'target_product'
}, inplace=True)

# Drop rows missing essential text or sentiment
tweet_clean.dropna(subset=['tweet_text', 'sentiment'], inplace=True)

# Replace missing target products with a meaningful label
tweet_clean['target_product'].fillna('Unknown_Product', inplace=True)

# -----------------------------
# 2. Remove duplicates
# -----------------------------
tweet_clean.drop_duplicates(inplace=True)

# -----------------------------
# 3. Clean sentiment labels
# -----------------------------
tweet_clean['sentiment'] = tweet_clean['sentiment'].replace({
    'No emotion toward brand or product': 'Neutral emotion',
    "I can't tell": None
})
tweet_clean.dropna(subset=['sentiment'], inplace=True)

# -----------------------------
# 4. Create simple text-based features
# -----------------------------
tweet_clean['tweet_length'] = tweet_clean['tweet_text'].str.len()
tweet_clean['word_count'] = tweet_clean['tweet_text'].str.split().apply(len)
```

## IV – Text Processing and Feature Engineering
---

In this step, we transform raw tweet text into features suitable for machine learning. We clean and standardize the text by removing noise like URLs, special characters, and stopwords, and convert it into numerical representations using **tokenization, lemmatization, and vectorization (e.g., TF-IDF)**.  

We also engineer features such as **tweet length, word count, and keyword presence** to enrich the dataset. These steps enable models to effectively learn patterns and accurately classify sentiment toward Apple and Google products.


```python
# -----------------------------
# IV – Text Processing & Feature Engineering
# -----------------------------

# Make a copy of the cleaned dataset
df_proc = tweet_clean.copy()

# -----------------------------
# 1. Text Cleaning Function
# -----------------------------
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    """
    Clean tweet text by:
    - Lowercasing
    - Removing URLs, mentions, hashtags, and punctuations
    - Removing stopwords
    - Lemmatization
    """
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # remove URLs
    text = re.sub(r"@\w+|#\w+", "", text)                # remove mentions/hashtags
    text = text.translate(str.maketrans("", "", string.punctuation))  # remove punctuation
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(words)

# Apply cleaning function
df_proc['clean_text'] = df_proc['tweet_text'].apply(clean_text)

# -----------------------------
# 2. Text Vectorization (TF-IDF)
# -----------------------------
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # limit features for efficiency
X_text = tfidf_vectorizer.fit_transform(df_proc['clean_text'])

# -----------------------------
# 3. Additional Engineered Features
# -----------------------------
# Features like tweet length, word count, mentions of Apple/Google
df_proc['apple_mention'] = df_proc['tweet_text'].str.contains('apple|ipad|iphone', case=False, regex=True).astype(int)
df_proc['google_mention'] = df_proc['tweet_text'].str.contains('google|android', case=False, regex=True).astype(int)

X_engineered = df_proc[['tweet_length', 'word_count', 'apple_mention', 'google_mention']].values

# -----------------------------
# 4. Combine Text Features and Engineered Features
# -----------------------------
from scipy import sparse as sp

# Combine TF-IDF text features with numeric features into one matrix
X_features = sp.hstack([X_text, sp.csr_matrix(X_engineered)])

# -----------------------------
# 5. Define Target Variable
# -----------------------------
y_target = df_proc['sentiment']
```

```python
# Verify Feature matrix shape
print("Feature matrix shape (X_features):", X_features.shape)
```

```python
# Verify target shape
print("Target variable shape (y_target):", y_target.shape)
```

```python
# Verify target value counts
print("\nSample target values:")
print(y_target.value_counts())
```

The dataset is now clean and preprocessed, with textual data transformed into a feature matrix and a clearly defined target variable. Most tweets are neutral, highlighting class imbalance. Engineered features like tweet length complement TF-IDF vectors. With everything prepared, we are ready to move on to **modeling** to predict tweet sentiment.

### 1. Multinomial Naive Bayes Model
---

We start with **Multinomial Naive Bayes**, a simple yet effective baseline for text classification.  
It uses word frequencies to predict tweet sentiment, providing a quick benchmark before exploring advanced NLP models.


```python
# -----------------------------
# Improved Multinomial Naive Bayes Model
# -----------------------------
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

# -----------------------------
# 1. Train-test split (stratified)
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_features, 
    y_target, 
    test_size=0.2, 
    random_state=42,
    stratify=y_target
)

# -----------------------------
# 2. Initialize & train MNB with smoothing
# -----------------------------
mnb_model = MultinomialNB(alpha=0.5)  # smoothing to improve generalization
mnb_model.fit(X_train, y_train)

# -----------------------------
# 3. Make predictions
# -----------------------------
y_pred = mnb_model.predict(X_test)
```

### Model Evaluation – Multinomial Naive Bayes
---

We assess the Multinomial Naive Bayes model using accuracy, precision, recall, and F1-score to understand its ability to classify **positive**, **negative**, and **neutral** tweets.


#### ***Evaluation metrics***
Bellow, we can notice that:  
**The overall accuracy has improved to around 76%, with a macro F1 score of 53%, reflecting better performance in capturing minority classes compared to the previous model.**  
These results highlight that combining TF-IDF text features with engineered features (like tweet length, word count, and mentions of Apple/Google) helped improve predictions for underrepresented sentiments. The increase in the macro F1 score indicates that the model is now less biased toward the majority class (neutral) and shows a more balanced ability to predict positive, negative, and neutral tweets.  


```python

# Accuracy, precision, recall, F1-score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (macro): {precision:.4f}")
print(f"Recall (macro): {recall:.4f}")
print(f"F1 Score (macro): {f1:.4f}")

# Detailed classification report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

```

- **Business context:**  

The model shows **moderate accuracy (66%)**, capturing general sentiment trends well. However, the **low macro recall (43%) and F1 score (42%)** indicate it struggles with **positive and negative tweets**.  

From a business perspective, this means important customer feedback—both praise and complaints—may be **missed**, highlighting the need for **better feature engineering or more advanced NLP models** to extract actionable insights.


#### ***Confusion Matrix***

The confusion matrix shows that most **neutral tweets** (middle row) are still correctly classified, while **negative tweets** (top row) are mostly misclassified as neutral, with only a few correctly predicted. **Positive tweets** (bottom row) are partially captured, but a large portion is also confused with neutral, reflecting that the model still exhibits bias toward the dominant neutral class.


```python
# Confusion Matrix with Counts & Percentages
cm = confusion_matrix(y_test, y_pred, labels=mnb_model.classes_)
cm_percent = cm / cm.sum() * 100

fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(
    cm,
    annot=np.array([f"{v}\n({p:.1f}%)" for v, p in zip(cm.flatten(), cm_percent.flatten())]).reshape(cm.shape),
    fmt='',
    cmap="PuBuGn",           # soft purple-blue-green
    cbar=False,
    linewidths=1,
    linecolor='white',
    annot_kws={"size": 12, "weight": "bold", "color": "black"}
)

# Labels and title
ax.set_xlabel("Predicted Label", fontsize=14, fontweight='bold')
ax.set_ylabel("True Label", fontsize=14, fontweight='bold')
ax.set_xticklabels(mnb_model.classes_, fontsize=12, fontweight='bold')
ax.set_yticklabels(mnb_model.classes_, fontsize=12, fontweight='bold')
plt.title("Confusion Matrix - MultinomialNB (Counts & %)", fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()

plt.savefig("Images/cmatrix.png", dpi=300, bbox_inches="tight", facecolor="white")

plt.show()
```
